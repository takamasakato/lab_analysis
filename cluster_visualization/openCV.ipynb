```python
#BGRの順番で保存し3つの画像を重ね合わせる

files=glob.glob('*.tif')
#img1 = cv2.imread('181109-aldocaGCaMP-6dpf-fish5-D20um-nostim1-TurboTransAcc-10000dFF-snapshots.ome0001.tif')
#img2 = cv2.imread('181109-aldocaGCaMP-6dpf-fish5-D20um-nostim1-TurboTransAcc-10000dFF-snapshots.ome0001.tif')

img1 = cv2.imread(files[0])
img2 = cv2.imread(files[0])

img1[:,:,:] = 0
img2[:,:,:] = 0

#gとbを初期化
img_gbr = cv2.split(img2)
g = img_gbr[2]
b = img_gbr[2]


for i, file_name in enumerate(files):
    img1[:,:,:] = 0
    img = cv2.imread(file_name,0)
    image, contours,hierarchy = cv2.findContours(img,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)#要注意(RETR_EXTERNAL:このフラグを使うと，全子要素は無視され最も外側の輪郭のみが返されます)cv2.CHAIN_APPROX_NONE : 全て。
    img_contour = cv2.drawContours(img1,contours,-1,(0,0,30),1)#要注意,なぜか引数であるimg1が変更されていた
    img2 = cv2.add(img2,img_contour)#確認済み

    
    
    if i % 100 == 0 and i % 200 != 0:
        #rとの合成とgの更新
        img_bgr = cv2.split(img2)
        r = img_bgr[2]
        #ひとつ
        img_merge = cv2.merge((g,b,r))
        img_RGB = cv2.cvtColor(img_merge, cv2.COLOR_BGR2RGB)
        plt.imshow(img_RGB)
        plt.show()
        #print('Hi')


        g = img_bgr[2]
        img2[:,:,:] = 0

    elif i % 100 == 0 and i % 200 == 0:
        #rとの合成とbの更新
        img_bgr = cv2.split(img2)
        r = img_bgr[2]
        img_merge = cv2.merge((b,g,r))
        img_RGB = cv2.cvtColor(img_merge, cv2.COLOR_BGR2RGB)
        plt.imshow(img_RGB)
        plt.show()

        b = img_bgr[2]
        img2[:,:,:] = 0
```

- matplotlibはRGBだが、opencvはBGRなので出力に注意
- 現在が赤色、一つ前が緑、二つ前が青となるようにした
- それぞれの単色とマージ像を示す。
- 過去二つよりも以前の状況と比較してみる
- 領域の伸び縮みも観察できるのでそこについても考察

```python
#BGRの順番で保存し3つの画像を重ね合わせる

files=glob.glob('*.tif')
#img1 = cv2.imread('181109-aldocaGCaMP-6dpf-fish5-D20um-nostim1-TurboTransAcc-10000dFF-snapshots.ome0001.tif')
#img2 = cv2.imread('181109-aldocaGCaMP-6dpf-fish5-D20um-nostim1-TurboTransAcc-10000dFF-snapshots.ome0001.tif')

img1 = cv2.imread(files[0])
img2 = cv2.imread(files[0])

img1[:,:,:] = 0
img2[:,:,:] = 0

#gとbを初期化
img_gbr = cv2.split(img2)
g = img_gbr[2]
b = img_gbr[2]


for i, file_name in enumerate(files):
    img1[:,:,:] = 0
    img = cv2.imread(file_name,0)
    image, contours,hierarchy = cv2.findContours(img,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    img_contour = cv2.drawContours(img1,contours,-1,(0,0,60),1)
    img2 = cv2.add(img2,img_contour)

    
    
    if i % 50 == 0 and i % 100 != 0:
        #rとの合成とgの更新
        img_bgr = cv2.split(img2)
        r = img_bgr[2]
        #ひとつ
        img_merge = cv2.merge((g,b,r))
        img_RGB = cv2.cvtColor(img_merge, cv2.COLOR_BGR2RGB)
        plt.imshow(img_RGB)
        plt.show()


        g = img_bgr[2]
        img2[:,:,:] = 0

    elif i % 50 == 0 and i % 100 == 0:
        #rとの合成とbの更新
        img_bgr = cv2.split(img2)
        r = img_bgr[2]
        img_merge = cv2.merge((b,g,r))
        img_RGB = cv2.cvtColor(img_merge, cv2.COLOR_BGR2RGB)
        plt.imshow(img_RGB)
        plt.show()

        b = img_bgr[2]
        img2[:,:,:] = 0
```

- 中心部では独立した活動が見られ、活動領域の重なり合いがすくない
- 外側では中心部と比較して活動領域が重なりあうことが多いように感じられる（活動の複雑さの違いなのかもしれない）
- 外側と内側では活動における周囲のプルキンエ細胞のふるまい方がことなるように感じられる。
